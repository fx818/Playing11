```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder

```

```python
data = pd.read_csv("all_season_batting_card.csv")

print(len(data))

data = data[:8700]

print(len(data))
```

```python
data = data.drop(['match_name', 'link','minutes', 'commentary', 'shortText', 'fullName', 'runningScore', 'city','home_team','away_team','current_innings'], axis=1)
```

```python
data['captain'] = data['captain'].apply(lambda x: 1 if x==True else 0)
data['isNotOut'] = data['isNotOut'].apply(lambda x: 1 if x==True else 0)
```

```python
data.head()
```

```python
le = LabelEncoder()
data['name'] = le.fit_transform(data['name'])
# print(data)
# data['name'][:120:5]
```

```python
data.head()
```

```python
le = LabelEncoder()
data['venue'] = le.fit_transform(data['venue'])
data['country'] = le.fit_transform(data['country'])
```

```python
data.head()
```

```python
# Before fitting we need to fill the missing data
data.isna().sum()
```

```python
data['runs'].fillna(0,inplace=True)
data['ballsFaced'].fillna(0,inplace=True)
# data['minutes'].fillna(0,inplace=True)
data['fours'].fillna(0,inplace=True)
data['sixes'].fillna(0,inplace=True)
data['strikeRate'].fillna(0,inplace=True)
data['runningOver'].fillna(0,inplace=True)

data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
len(data)
```

```python
data['strikeRate'] = data['strikeRate'].replace('-',0)
```

```python
# Making out the features and label column

x = data.drop('strikeRate',axis=1)
y = data['strikeRate']
```

```python
x.head()
```

```python
# Splitting the data

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
# Choosing the right model
# As we are going to predict the value so we use linearRegression

from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
```

```python
len(data)
```

```python
# Fit the model
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
